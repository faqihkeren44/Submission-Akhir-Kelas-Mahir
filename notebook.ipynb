## Data Prepocessing

**Menghapus feature yang tidakk mempengaruhi data**

Dari semua feature/kolom, sebagian tidak mempengaruhi status siswa atau mempengaruhi data yang mempengaruhi status. Agar model yang dibuat dapat memprediksi lebih baik, feature/kolom ini akan kita hapus. Selanjutnya untuk mengecek apakah sudah terhapus digunakan kode `info()`

```python
model_df = main_df.copy()
model_df = model_df.drop(columns=['Daytime_evening_attendance',
                          'Curricular_units_2nd_sem_evaluations',
                          'Fathers_occupation', 'Mothers_occupation',
                          'Fathers_qualification', 'Mothers_qualification',
                          'Curricular_units_1st_sem_evaluations',
                          'Curricular_units_1st_sem_without_evaluations',
                          'Curricular_units_2nd_sem_without_evaluations',
                          'Curriculum_difference', 'Application_mode',
                          'Application_order', 'Course',
                          'Previous_qualification', 'Admission_grade',
                          'Previous_qualification_grade'], axis=1)
```

```python
model_df.info()
```

**Menghapus data yang tidak dibutuhkan**

Karena model yang dibuat akan menganalisis siswa akan dropout atau tidak, maka kolom status perlu diubah terlebih dahulu, yaitu untuk siswa yang dropout bernilai 1, dan yang tidak bernilai 0. Lalu kolom status yang asli dihapus.

```python
model_df['dropout'] = model_df['Status'].map(lambda x: 1 if x == 0 else 0)
model_df.drop(['Status'], axis=1, inplace=True)
model_df['dropout']
```

## Modeling

**Import library yang dibutuhkan**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC,SVC
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
```

**Membagi dataset**

Feature yang dipakai yaitu feature yang sudah dipilih dan disimpan pada model_df.

```python
X = model_df.drop('dropout', axis=1)
y = model_df['dropout']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

**Scaler**

Untuk menangani data bias yang ada, python menyediakan `StandardScaler()` dan `MinMaxScaler()`

```python
# scaler=StandardScaler()
# scaler.fit(X_train)
# X_train=pd.DataFrame(scaler.transform(X_train),columns=X_train.columns,index=X_train.index)
# X_test=pd.DataFrame(scaler.transform(X_test),columns=X_test.columns,index=X_test.index)

minmaxscaler = MinMaxScaler()
minmaxscaler.fit(X_train)
X_train=pd.DataFrame(minmaxscaler.transform(X_train),columns=X_train.columns,index=X_train.index)
X_test=pd.DataFrame(minmaxscaler.transform(X_test),columns=X_test.columns,index=X_test.index)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

**Pembuatan Model**

Untuk mengetahui model apa yang memiliki tingkat akurasi paling tinggi, semua jenis model dicoba satu persatu, lalu ditamplikan tingkat akurasinya dengan kode `print()`

```python
models={'LogisticRegression()':LogisticRegression(),
        'K-Nearest Neighbors': KNeighborsClassifier(),
        'Decision Tree':DecisionTreeClassifier(),
        'Support Vector Machine(Linear Kernel)':LinearSVC(),
        'Support Vector Machine(Non-Linear Kernal)':SVC(),
        'Neural Network':MLPClassifier(),
        'Random Forest':RandomForestClassifier(),
        'Gradient Boosting':GradientBoostingClassifier()}
```

```python
for name, model in models.items():
    print(name)
    model.fit(X_train,y_train)
    print(model.score(X_test,y_test))
```

Dari daftar model di atas, tingkat accuracy paling tinggi ada pada model Gradient Boosting. Maka `GradientBoostingClassifier()` yang dipilih menjadi model.

```python
model = GradientBoostingClassifier()
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

## Evaluation

Untuk mengetahui tingkat akurasi dari model yang telah dibuat, dapat ditampilkan dengan berbagai cara. Yang paling simple yaitu dengan mengguanakan kode `accuracy_score(y_test, pred)`. Semakin tinggi nilainya, semakin baik akurasi model

```python
accuracy_score(y_test, pred)
```

Mengecek tingkat akurasi juga bisa dilakukan dengan menggunakan kode `confusion_matrix()`. Evaluasi ini akan menghasilkan nilai yang benar diprediksi dan yang salah diprediksi oleh model. Hasil evaluasi ini dapat juga divisualisasikan menggunakan matplotlib

Pada nilai 0 dan 1 vertikal menunjukan nilai sebenarnya pada data, sedangkan nilai 0 dan 1 horizontal adalah nilai yang diprediksi oleh model. Jika nilai seharusnya 0 dan model memprediksi 0, berarti model memprediksi hasil yang sesuai.

```python
matrix = confusion_matrix(y_test, pred)
matrix
```

```python
matrix = confusion_matrix(y_test, pred)
plt.matshow(matrix)
plt.colorbar()

plt.title('Confusion Matrix')
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```
